2.- Actualizar a valor num√©rico la variable Sex

```python
#otro codigo para cambiar valores de sexo 1,0
#passenger['sex'].replace['female','male'],[0,1], inplace=True
pd.get_dummies(passengers['Sex'])#extraemos los valores de la variable Sex
```

```python
gender=pd.get_dummies(passengers['Sex'],drop_first=True)#creamos una nueva columna gender y eliminamos de la columna Sex la variable female
```

```python
passengers['sex']=gender#cambiamos el nombre de la columna gender por sex
```

```python
# Eliminamos todas las columnas que no necesitaremos
passengers.drop(['PassengerId','SibSp','Parch', 'Ticket', 'Fare','Cabin', 'Embarked','Sex'],axis=1, inplace=True)
```

```python
passengers.drop(['Name'],axis=1, inplace=True)
```

- 3.- Rellenamos los valores vac√≠os de la variable Age con su Media

```python
passengers['Age'].values
```

```python
passengers.isnull().sum()
```

```python
#RELLENAMOS VALORES DE  LA VARIABLE Age POR SU MEDIA QUE ES 30
passengers.describe()
```

```python
passengers['Age']= passengers['Age'].replace(np.nan, 30)
passengers.head(22)
```

```python
passengers.isnull().sum() #OBSERVAMOS QUE YA NO TENEMOS VALORES NAN EN LA VARIABLE Age
```

- 4.- Creamos la columna FirstClass

```python
 #utilicemos la columna Pclass, o la clase de pasajeros, como otra caracter√≠stica. Cree una nueva
#columna llamada FirstClass que almacene 1 para todos los pasajeros en
#primera clase y 0 para todos los dem√°s pasajeros.
passengers['FirstClass']= np.where(passengers['Pclass']==1,1,0)
passengers.head(10)
```

- 5.- Creamos la columna SecondClass

```python
#Cree una nueva columna llamada SecondClass que almacene 1 para todos
#los pasajeros en segunda clase y 0 para todos los dem√°s pasajeros.

passengers['SecondClass']= np.where(passengers['Pclass']==2,1,0)
passengers.head(10)
```

- 6.- Creamos las variables features y survival

```python
features=passengers[['Age','sex','FirstClass','SecondClass']]
survival=passengers['Survived']
```

```python
features
```

- 7.- Dividimos los datos con el m√©todo train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
features_train, features_test, survival_train, survival_test= train_test_split(features,survival, test_size=0.33, random_state=101)
```

- 8 y 9 Utilizamos la regularizaci√≥n y creamos un modelo de Regresi√≥n Log√≠stica

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
lr=LogisticRegression()
```

```python
lr.fit(features_train, survival_train)
```

```python
#PRECISION ALGORITMO REGRESION LOGISTICA
predict=lr.predict(features_test)
print('Precisi√≥n de la Regresi√≥n Log√≠stica: ')
print(lr.score(features_train,survival_train))
```

- Adicionalmente se aplicaron algoritmos de Soporte de Vectores y Vecinos m√°s Cercanos para una comparaci√≥n m√°s profunda

```python
#PRECISION ALGORITMO SOPORTE DE VECTORES
svc=SVC()
svc.fit(features_train,survival_train)
Y_pred=svc.predict(features_test)
print('Precisi√≥n de Soporte de Vectores: ')
print(svc.score(features_train,survival_train))
```

```python
#PRECISION ALGORITMO VECINOS MAS CERCANOS
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(features_train,survival_train)
print('Precisi√≥n de Vecinos m√°s cercanos: ')
print(knn.score(features_train,survival_train))
```

```python
#ANALISIS DE LOS ALGORITMOS UTILIZADOS
print('Precisi√≥n de la Regresi√≥n Log√≠stica: ')
print(lr.score(features_train,survival_train))
print('')
print('Precisi√≥n de Soporte de Vectores: ')
print(svc.score(features_train,survival_train))
print('')
print('Precisi√≥n de Vecinos m√°s cercanos: ')
print(knn.score(features_train,survival_train))
```

- Por lo que podemos concluir que el mejor algoritmo es el de Vecinos m√°s Cercanos con 84% de confianza

- 10, 11 y 12 Se obtiene el puntaje de entrenamiento, la puntuaci√≥n de prueba y los coeficientes de caracter√≠stica determoinados por el modelo.
